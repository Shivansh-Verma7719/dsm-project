# 02 — Why Don't More People Invest? Behaviour & Barriers

SEBI Household Survey 2015. Urban India.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

DATA = Path("..") / "data"
SEBI = DATA / "sebi"

plt.rcParams.update({"figure.dpi": 130, "axes.spines.top": False, "axes.spines.right": False})
```

## 1. Awareness Gap — Investor vs Non-Investor (File 15)

```python
aw = pd.read_csv(SEBI / "15_urban_household_awareness_investment_instruments.csv")
aw.columns = ["country", "year", "group", "mf", "equity", "debentures", "derivatives", "commodity"]

print(aw[["group", "mf", "equity", "debentures", "derivatives", "commodity"]].to_string(index=False))

instruments = ["mf", "equity", "debentures", "derivatives", "commodity"]
labels = ["Mutual Funds", "Equity", "Debentures", "Derivatives", "Commodity Futures"]

fig, ax = plt.subplots(figsize=(8, 4))
x = range(len(instruments))
w = 0.35
for i, row in aw.iterrows():
    offset = -w/2 if i == 0 else w/2
    ax.bar([v + offset for v in x], row[instruments].values, w, label=row["group"])
ax.set_xticks(list(x))
ax.set_xticklabels(labels, rotation=20, ha="right")
ax.set_ylabel("Awareness (%)")
ax.set_title("Awareness of Investment Instruments: Investor vs General Urban (2015)")
ax.legend()
plt.tight_layout()
plt.show()
```

## 2. Barriers to Participation (File 10)

```python
bar = pd.read_csv(SEBI / "10_reasons_for_low_participation_securities_markets.csv")
bar.columns = ["country", "year", "rank", "reason", "n"]

top = bar.groupby("reason")["n"].sum().sort_values(ascending=True)
print(top.to_string())

fig, ax = plt.subplots(figsize=(8, 5))
top.plot(kind="barh", ax=ax, color="tomato")
ax.set_xlabel("Total votes across ranks")
ax.set_title("Barriers to Securities Market Participation (Urban India, 2015)")
plt.tight_layout()
plt.show()
```

## 3. Motivations to Invest (File 09)

```python
mot = pd.read_csv(SEBI / "09_reasons_for_clients_to_invest_securities_markets.csv")
mot.columns = ["country", "year", "rank", "reason", "n"]

top_m = mot.groupby("reason")["n"].sum().sort_values(ascending=True)
print(top_m.to_string())

fig, ax = plt.subplots(figsize=(8, 5))
top_m.plot(kind="barh", ax=ax, color="steelblue")
ax.set_xlabel("Total votes across ranks")
ax.set_title("Motivations to Invest in Securities Markets (Urban India, 2015)")
plt.tight_layout()
plt.show()
```

## 4. Information Sources — Formal vs Informal (File 06)

```python
info = pd.read_csv(SEBI / "06_information_sources_market_participants.csv")
info.columns = ["country", "year", "source", "rank", "pct"]

top_info = info.groupby("source")["pct"].sum().sort_values(ascending=True)
print(top_info.to_string())

fig, ax = plt.subplots(figsize=(7, 5))
top_info.plot(kind="barh", ax=ax, color="seagreen")
ax.set_xlabel("Summed rank-weighted %")
ax.set_title("Information Sources Used by Market Participants (2015)")
plt.tight_layout()
plt.show()
```

## 5. Risk/Return Perception (File 07)

```python
rr = pd.read_csv(SEBI / "07_perceived_risk_return_time_horizon.csv")
rr.columns = ["country", "year", "instrument", "risk_score", "return_score", "time_score"]
print(rr[["instrument", "risk_score", "return_score", "time_score"]].to_string(index=False))

fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(rr["risk_score"], rr["return_score"], s=80, color="steelblue")
for _, row in rr.iterrows():
    ax.annotate(row["instrument"], (row["risk_score"], row["return_score"]),
                textcoords="offset points", xytext=(5, 3), fontsize=8)
ax.set_xlabel("Perceived Risk (mean score)")
ax.set_ylabel("Perceived Return (mean score)")
ax.set_title("Risk vs Return Perception by Instrument (Urban Investors, 2015)")
plt.tight_layout()
plt.show()
```

## 6. Investor Education Programme Exposure (File 11)

```python
iep = pd.read_csv(SEBI / "11_investor_education_program_participation_tabulation.csv")
iep.columns = ["country", "year", "investment", "non_participant", "pct_non", "participant", "pct_part", "total"]
print(iep[["investment", "non_participant", "participant", "total"]].to_string(index=False))

fig, ax = plt.subplots(figsize=(7, 4))
x = range(len(iep))
w = 0.35
ax.bar([i - w/2 for i in x], iep["non_participant"], w, label="Non-participant", color="tomato")
ax.bar([i + w/2 for i in x], iep["participant"],     w, label="IEP Participant",  color="steelblue")
ax.set_xticks(list(x))
ax.set_xticklabels(iep["investment"], rotation=20, ha="right")
ax.set_ylabel("Number of investors")
ax.set_title("IEP Exposure by Current Investment Type (2015)")
ax.legend()
plt.tight_layout()
plt.show()
```

## 7. Risk Mitigation Strategies (File 08)

```python
rm = pd.read_csv(SEBI / "08_risk_mitigation_process_urban_investors.csv")
rm.columns = ["country", "year", "rank", "known_company", "mutual_fund", "observe_market", "established_broker", "small_parts"]

strategies = ["known_company", "mutual_fund", "observe_market", "established_broker", "small_parts"]
labels = ["Known Company", "Mutual Fund", "Observe Market", "Established Broker", "Small Parts"]

weighted = rm[strategies].mul(rm["rank"].max() + 1 - rm["rank"], axis=0).sum()
weighted = weighted.sort_values(ascending=True)

print(weighted.to_string())
fig, ax = plt.subplots(figsize=(7, 4))
weighted.plot(kind="barh", ax=ax, color="steelblue")
ax.set_xlabel("Rank-weighted score")
ax.set_title("Risk Mitigation Strategies (Urban Investors, 2015)")
plt.tight_layout()
plt.show()
```

## 8. Holding Duration — Equities (IPO) & Mutual Funds (Files 16, 18)

```python
ipo = pd.read_csv(SEBI / "16_holding_period_equities_purchased_via_ipo.csv")
ipo.columns = ["country", "year", "period", "n", "pct", "cumulative"]
ipo = ipo.sort_values("pct", ascending=False)

mf = pd.read_csv(SEBI / "18_mutual_fund_holding_duration.csv")
mf.columns = ["country", "year", "period", "pct", "cumulative"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(range(len(ipo)), ipo["pct"], color="tomato")
ax1.set_xticks(range(len(ipo)))
ax1.set_xticklabels(ipo["period"], rotation=30, ha="right", fontsize=8)
ax1.set_ylabel("% of investors")
ax1.set_title("Equity (IPO) Holding Duration")

ax2.bar(range(len(mf)), mf["pct"], color="steelblue")
ax2.set_xticks(range(len(mf)))
ax2.set_xticklabels(mf["period"], rotation=30, ha="right", fontsize=8)
ax2.set_ylabel("% of investors")
ax2.set_title("Mutual Fund Holding Duration")

plt.tight_layout()
plt.show()
```
